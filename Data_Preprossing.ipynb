```python
import cv2, numpy as np
import dlib
import matplotlib.pyplot as plt
```

```python
face_detector = dlib.get_frontal_face_detector()
dlib_facelandmark = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
def augemented_images(image):
    brightness = np.random.uniform(0.5, 2.0)
    contrast = np.random.uniform(0.5, 2.0)
    adjusted_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)
    return [adjusted_image, cv2.flip(image,1)]
```

```python
def frames_extraction(path):
    cap = cv2.VideoCapture('original_sequences/youtube/c23/videos/'+path+'.mp4')
    ind = 1
    while cap.isOpened():
        ret,frame = cap.read()
        if not ret:
            break
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        obj = face_detector(grey)
        if len(obj)!=0:
            face = obj[0]
            img = frame[face.top():face.bottom(),face.left():face.right()]
            aug = augemented_images(img)
            cv2.imwrite(f"Images/Real/{path}_{ind}_0.png", img) 
            cv2.imwrite(f"Images/Real/{path}_{ind}_1.png", aug[0])
            cv2.imwrite(f"Images/Real/{path}_{ind}_2.png", aug[1])
        ind+=1
    print(f"{path} is completed")
```

```python
def frames_extraction_1(path):
    cap = cv2.VideoCapture(f'original_sequences/youtube/c23/videos/{path}.mp4')
    frames = []
    while cap.isOpened():
        ret,frame = cap.read()
        if not ret:
            break
        frames.append(frame)
    for i in np.random.choice(range(len(frames)),50, replace=False):
        frame = frames[i].copy()
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        obj = face_detector(grey)
        if len(obj)!=0:
            face = obj[0]
            img = frame[face.top():face.bottom(),face.left():face.right()]
            if len(img)!= 0:
                aug = augemented_images(img)
                cv2.imwrite(f"Images/Real/{path}_{i}_0.png", img) 
                cv2.imwrite(f"Images/Real/{path}_{i}_1.png", aug[0])
                cv2.imwrite(f"Images/Real/{path}_{i}_2.png", aug[1])
    print(f"{path} is completed")
```

```python
def frames_extraction_2(path, img_path, flag):
    if flag ==1:

        cap = cv2.VideoCapture(f'original_sequences/youtube/c23/videos/{path}.mp4')
    else:
        cap = cv2.VideoCapture(f'manipulated_sequences/Deepfakes/c23/videos/{path}.mp4')
    frames = []
    while cap.isOpened():
        ret,frame = cap.read()
        if not ret:
            break
        frames.append(frame)
    for i in np.random.choice(range(len(frames)),2, replace=False):
        frame = frames[i].copy()
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        obj = face_detector(grey)
        if len(obj)!=0:
            face = obj[0]
            img = frame[face.top():face.bottom(),face.left():face.right()]
            if len(img)!= 0:
                aug = augemented_images(img)
                cv2.imwrite(f"{img_path}/{path}_{i}_0.png", img) 
                cv2.imwrite(f"{img_path}/{path}_{i}_1.png", aug[0])
                cv2.imwrite(f"{img_path}/{path}_{i}_2.png", aug[1])
    print(f"{path} is completed")
```

```python
import os

real = os.listdir('original_sequences/youtube/c23/videos')[:100]
fake = os.listdir('manipulated_sequences/Deepfakes/c23/videos')[:100]

for i in real:
    frames_extraction_2(i.split('.')[0], 'Siamese Images/Real', 1)

for j in fake:
    frames_extraction_2(j.split('.')[0], 'Siamese Images/Fake', 0)
```
