```python
from tensorflow.keras.applications.xception import Xception, preprocess_input
from tensorflow.keras.models import Model, load_model
from tensorflow.keras import layers, losses, metrics, optimizers
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, LeakyReLU, Softmax
from tensorflow.keras.optimizers import SGD
```

```python
classfication_model = load_model('myModel')
```

```python
import cv2, numpy as np
import dlib
import matplotlib.pyplot as plt
face_detector = dlib.get_frontal_face_detector()
dlib_facelandmark = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
def video_classifiaction(video_path, model):
    cap = cv2.VideoCapture(video_path)
    real,fake = 0,0
    while cap.isOpened():
        ret,frame = cap.read()
        if not ret:
            break
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        obj = face_detector(grey)
        if len(obj)!=0:
            try:
                face = obj[0]
                my_image = frame[face.top():face.bottom(),face.left():face.right()]
                my_image = cv2.resize(my_image, (71,71), interpolation = cv2.INTER_AREA)
                my_image = my_image.reshape((1, my_image.shape[0], my_image.shape[1], my_image.shape[2]))
                my_image = preprocess_input(my_image)

                #make the prediction
                prediction = model.predict(my_image)
                if prediction[0][0] > 0.5:
                    real+=1
                else:
                    fake +=1
            except:
                continue
    return 1 if real > fake else 0
```

```python
import os
fake = [['manipulated_sequences/Deepfakes/c23/videos/'+i, 0] for i in sorted(os.listdir('manipulated_sequences/Deepfakes/c23/videos'))]
```

```python
result = open('result.txt','a')

for i in fake[0:50]:
  print(f"{i[0]} has started")
  k = video_classifiaction(i[0], classfication_model)
  result.write(f'{k}\n')
result.close()
```

```python
real_pred = [int(i) for i in open('real.txt','r').readlines()] 
fake_pred = [int(i) for i in open('fake.txt','r').readlines()] 
```

```python
pred = real_pred + fake_pred
```

```python
label = [1] * len(real_pred) + [0] * len(fake_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(label,pred)
```
